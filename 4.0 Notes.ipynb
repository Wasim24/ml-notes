# How does regularization help handle overfitting?
Overfitting implies high variance. If we control and contain the model parameters, then even if there's some noise in the data, it's impact will be way less. This is how controlling model parameters will help in controlling the variance. However, regularization will add little bias oo the model.
# What are the solutions to handle Overfitting?
- Provide more training data
- choose simpler model instead of complex one with lots of parameters
- Dimensionality reduction
    - feature selection: 
        - Example: SBS etc 
        - Note: reducing #features decreases curse of dimensionality and hence model will perform better on validation set]
        - check out http://scikit-learn.org/stable/modules/feature_selection.html for more
    - feature extraction:  we derive information from feature set to construct a new feature subspace. Ex: PCA, LDA etc
- Regularization

# why L1 leads to more sparse results than L2?
L1 & L2 regularization add constraints to the optimization problem. The curve H0 is the hypothesis. The solution to this system is the set of points where the H0 meets the constraints. Now, in the case of L2 regularization, in most cases, the the hypothesis is tangential to the ||w||_2. The point of intersection has both x1 and x2 components. On the other hand, in L1, due to the nature of ||w||_1, the viable solutions are limited to the corners, which are on one axis only - in the above case x1. Value of x2 = 0. This means that the solution has eliminated the role of x2 leading to sparsity. See the below 4 figures to understand the same in linear case and higher dimensions case.

# Feature Selection: Sequential Backward Selection (SBS):
SBS sequentially removes features from the full feature subset until the new feature subspace contains the desired number of features. In order to determine which feature is to be removed at each stage, we need to define criterion function <b>J </b> that we want to minimize. <b>J </b> can simply be the difference in performance of the classifier after and before the removal of a particular feature.
