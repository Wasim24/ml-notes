```python
import numpy as np
import matplotlib.pyplot as plt

class RandomDataGenerator:
    
    @staticmethod
    # Generates random data with xor of sample_size size. (defaulted to generate 200 samples)
    def use_xor(sample_size=200):
        np.random.seed(0)
        X_xor = np.random.randn(200,2)
        y_xor = np.logical_xor(X_xor[:,0]>0, X_xor[:,1]>0)
        y_xor = np.where(y_xor, 1, -1)
        # plot generated data to see the distribution
        plt.title('Random data of size 200 generated using XOR\n')
        plt.scatter(X_xor[y_xor==1, 0], X_xor[y_xor==1, 1], marker='x', c='b', label='1')
        plt.scatter(X_xor[y_xor==-1,0], X_xor[y_xor==-1, 1], marker='s', c='r', label='-1')
        plt.ylim(-3.0)
        plt.legend()
        plt.show()
        return X_xor, y_xor
```
