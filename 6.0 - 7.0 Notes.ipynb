# Why K-FOLD?
test_train_split will split the data into 2 parts. If we are not satisfied with the model, we would go retrain the same model with different params. But, if the model sees the same data again and again, over-fitting can happen. So, k-fold. Here, we split randomly as shown in the pic below.

# cross_val_score
In k-fold, we have to manually fit the split data and keep track of scores. cross_val_score is an add on top of k-fold where it takes estimator, num_folds, scoring module and returns all the scores.

# 1. K-Fold: 
- Idea is to not use test data to validate the performance when we want to try & figure out our model with different param values (Ex: LR with C = 0.1, LR with C=0.01, LR with C=10 etc)
- Reusing test data again and again may cause overfitting.
- So, k-fold came in.. We split training data into training & validation folds. 
- Iteratively fit model with different training and test folds as shown in the picture below.
- This way we are totally isolating Test data and run k-fold with different model params and pick the best one
- <b>Stratified K-Fold:</b> Compared to standard k-fold, in stratified, the distribution of class labels is maintained in each of the training folds. So, this is better than standard k-fold (where we just pik random datasets in training folds without replacement).
